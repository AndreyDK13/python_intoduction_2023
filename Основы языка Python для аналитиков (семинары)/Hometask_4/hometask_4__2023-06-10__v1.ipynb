## Условие 3: 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.  
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

* *карта корреляции цены с остальными атрирбутами*

```python
import numpy as np

corr_matrix = df.corr()[['price']]
corr_matrix = corr_matrix.sort_values('price',ascending = False)
# убираем цену и другие лишние атрибут, которые зависимы от аналогичных (например, sqft_above - составяная часть sqft_living и тп)
corr_matrix = corr_matrix.drop(['price', 'sqft_above', 'sqft_basement', 'lat', 'long', 'id', 'sqft_lot15'])
plt.figure(figsize=(5,10))
sns.heatmap(corr_matrix, vmin = -1, vmax = +1, annot=True, linewidths=0.5, cmap='coolwarm')
```

> **ВЫВОД:** наибольшее влияние на цену оказывает площадь, оценка, площадь жилья в округе, к-во ванных комнат, вид, к-во спален, вид на побережье, этажность  
Далее тут можно было бы не исследовать, но по заданию нужно 5 графиков

* *Отдельно поссиледуем среднюю цену продаж в зависимости от года*

```python
df['year_of_sale'] = df['date'].apply(lambda x: x.split('-')[0])
plt.figure(figsize=(10,6))
sns.boxplot(x=df['year_of_sale'], y=df['price'], whis=1.5)
```

> Как видим, в среднем цены практически не менялись, но сделок с завышенным ценником в 2014г было значительно больше, чем в 2015

* *Исследуем более детально оценку (grade)*

```python
plt.figure(figsize=(10,6))
sns.boxplot(x=df['grade'], y=df['price'], whis=1.5)
```

> Прослеживается явная экспоненциальная зависимость цены дома от оценки, при этом в оценках >5 было много продаж по завышенной стоимости

* *Исследуем более детально к-во этажей*

```python
plt.figure(figsize=(10,6))
sns.boxplot(x=df['floors'], y=df['price'], whis=1.5)
```

> Как видно из график, прослеживается аналогичная тенденция: рост цены в зависимости от к-ва этажей, при этом 3+ этажа ценятся меньше, чем 2,5-  
Топовым считаются дома с 2 этажами и подвалом - средняя существенно выше других и разброс цен также выше
